```python
import plotly.graph_objs as go 
import sys
import paho.mqtt.client
import json
```

```python

graf_n=go.FigureWidget()

graf_n 
```

```python
graf_o=go.FigureWidget()
graf_o
```

```python
graf_a=go.FigureWidget()
graf_a
```

```python
graf_p=go.FigureWidget()

graf_p
```

```python
def complet():
    

    graf_n.add_scatter(x=[],y=[])
    graf_o.add_scatter(x=[],y=[])
    graf_a.add_scatter(x=[],y=[])
    graf_p.add_scatter(x=[],y=[])
    TempN=[]
    TimeN=[]
    TempO=[]
    TimeO=[]
    TempA=[]
    TimeA=[]
    CantP=[]
    TimeP=[]

    def on_connect(client,userdata,flags,rc):
        print('connected (%s)' % client._client_id)
        client.subscribe(topic='casa/#', qos=2)

    def on_message(client,userdata,message):
        print('-----------------------------')
        print('topic: %s' % message.topic)
        if(message.topic =='casa/cocina/nevera'):
            print('--NEVERA--')
            m_decode=str(message.payload.decode("utf-8","ignore"))
            m_in=json.loads(m_decode)
            #print(m_in)
            TempN.append(m_in['tem_neve'])
            TimeN.append(m_in["tim_neve"])
            
            graf_n.data[0].x=TimeN
            graf_n.data[0].y=TempN
        elif(message.topic =='casa/cocina/olla'):
            print('--OLLA--')
            m_decode=str(message.payload.decode("utf-8","ignore"))
            m_in=json.loads(m_decode)
            #print(m_in)
            TempO.append(m_in['tem_olla'])
            TimeO.append(m_in["time"])
            print(TimeO[0])
            graf_o.data[0].x=TimeO
            graf_o.data[0].y=TempO
        elif(message.topic =='casa/sala/alexa'):
            print('--ALEXA--')
            m_decode=str(message.payload.decode("utf-8","ignore"))
            m_in=json.loads(m_decode)
            #print(m_in)
            TempA.append(m_in['temp'])
            TimeA.append(m_in["time"])
            
            graf_a.data[0].x=TimeA
            graf_a.data[0].y=TempA
        elif(message.topic =='casa/sala/persona'):
            print('--PERSONA--')
            m_decode=str(message.payload.decode("utf-8","ignore"))
            m_in=json.loads(m_decode)
            #print(m_in)
            CantP.append(m_in['cant'])
            TimeP.append(m_in["time"])
            
            graf_a.data[0].x=TimeP
            graf_a.data[0].y=CantP
            


    def main():
        client = paho.mqtt.client.Client(client_id='cocina grafico', clean_session=False)
        client.on_connect = on_connect
        client.on_message = on_message
        client.connect(host='127.0.0.1', port=1883)
        client.loop_forever()


    if __name__ == '__main__':
        main()
        sys.exit(0)
```
